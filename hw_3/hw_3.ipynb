Выведите самое популярное мужское и самое популярное женское имя на корабле

```python
import re
```

```python
def name(full): # сложноооо
  n_match = re.search(r'\(([^)]+)\)', full)
  if n_match:
    return n_match.group(1).split()[0]
  else:
    tokens = full.replace('.', '').split()
    for i, token in enumerate(tokens):
      if token in ['Mr', 'Miss', 'Mrs', 'Master']:
        if i + 1 < len(tokens):
          return tokens[i+1]
    return None
```

```python
df['FirstName'] = df['Name'].apply(name)
```

```python
n_male = df[df['Sex'] == 'male']['FirstName'].value_counts().idxmax()
n_female = df[df['Sex'] == 'female']['FirstName'].value_counts().idxmax()

print(f'Male name: {n_male}; Female name: {n_female}')
```

Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека)

```python
cabin_all = df['Cabin'].value_counts().to_frame()

df_cabin_all = pd.DataFrame(cabin_all)
df_cabin_all = df_cabin_all.reset_index()
df_cabin_all.columns = ['Cabin', 'Counts']

cabin_n = df_cabin_all[df_cabin_all['Counts'] > 1]
cabin_n_final = cabin_n['Cabin'].count()
print(f'Количество n-местных кабин: {cabin_n_final}')
```
