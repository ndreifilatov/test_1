```python
class Chocolate:
  def __init__ (self, price: float, add: str = None):
    self.price = price
    self.add = add

    if self.price <= 0:
     raise ValueError('Цена меньше нуля')

  def show_my_chocolate(self):
    if self.add == None:
      print('Обычный шоколадный батончик')
    else:
      print('Шоколадный батончик, добавка:', self.add)

  def __str__(self):
    return f'Шоколадный батончик, {self.price} рублей'
```

```python
chocolate_1 = Chocolate(69.69)
print(chocolate_1)

chocolate_1.show_my_chocolate()
```

```python
chocolate_2 = Chocolate (77.77, 'фундук')
print(chocolate_2)

chocolate_2.show_my_chocolate()
```

```python
chocolate_3 = Chocolate (-1.5, 'марципан')
print(chocolate_3)

chocolate_3.show_my_chocolate()
```
