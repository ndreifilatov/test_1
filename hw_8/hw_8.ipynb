```python
import requests
import pandas as pd
```

#Работа с API

##Получить информацию обо всех книгах и сохранить в pd.DataFrame

```python
api_base_url = 'https://www.anapioficeandfire.com/api'

books_list = requests.get(
    f'{api_base_url}/books'
).json()
```

```python
df_books = pd.DataFrame(books_list)
df_books.head()
```

##Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame

```python
houses_list = requests.get(
    f'{api_base_url}/houses'
).json()

df_houses = pd.DataFrame(houses_list)
df_houses.head()
```

##Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame

```python
houses_words_list = requests.get(
    f'{api_base_url}/houses?hasWords=True'
).json()

df_houses_words_list = pd.DataFrame(houses_words_list)
df_houses_words_list.head()
```
