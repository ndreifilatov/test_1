##Посчитайте средний возраст пассажиров и его стандартное отклонение с помощью bottleneck

```python
import bottleneck as bn
```

```python
age = df['Age']

print(f'Средний возраст пассажиров: {bn.nanmean(age)}')
print(f'Стандартное отклонение: {bn.nanstd(age)}')
```

##Считайте датасет из файла Housing.csv

##Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных - напишите свои выводы в комментариях

```python
print(df_2.dtypes)
```

```python
print(f"price, max: {df_2['price'].max()}; min: {df_2['price'].min()}")
print(f"area, max: {df_2['area'].max()}; min: {df_2['area'].min()}")
print(f"bedrooms, max: {df_2['bedrooms'].max()}; min: {df_2['bedrooms'].min()}")
print(f"bathrooms, max: {df_2['bathrooms'].max()}; min: {df_2['bathrooms'].min()}")
print(f"stories, max: {df_2['stories'].max()}; min: {df_2['stories'].min()}")
print(f"parking, max: {df_2['parking'].max()}; min: {df_2['parking'].min()}")
```

в соответствии с данными о максимальных и минимальных значениях выше, price оптимизируем в int32, area оптимизируем в int16, остальные целочисленнные данные оптимизируем в int8

для оптимизации данных object следует перейти к категориальным значениям, так как используются повторяющиеся однотипные элементы

##Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации

```python
df_3 = pd.read_csv('Housing.csv')

df_3['price'] = df_3['price'].astype(np.int32)
df_3['area'] = df_3['area'].astype(np.int16)
df_3['bedrooms'] = df_3['bedrooms'].astype(np.int8)
df_3['bathrooms'] = df_3['bathrooms'].astype(np.int8)
df_3['stories'] = df_3['stories'].astype(np.int8)
df_3['parking'] = df_3['parking'].astype(np.int8)

df_3['mainroad'] = df_3['mainroad'].astype('category')
df_3['guestroom'] = df_3['guestroom'].astype('category')
df_3['basement'] = df_3['basement'].astype('category')
df_3['hotwaterheating'] = df_3['hotwaterheating'].astype('category')
df_3['airconditioning'] = df_3['airconditioning'].astype('category')
df_3['prefarea'] = df_3['prefarea'].astype('category')
df_3['furnishingstatus'] = df_3['furnishingstatus'].astype('category')
```
